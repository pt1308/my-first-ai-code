```python
dates = ["3月17日", "3月18日", "3月19日", "3月23日", "3月24日","3月25日","3月27日"]
pace = [5.28,5.23,5.53,5.36,5.26,4.56,5.52]
import matplotlib.pyplot as plt

plt.plot(dates, pace, marker='o')
plt.title("我的跑步配速进步曲线")
plt.xlabel("日期")
plt.ylabel("配速（分钟/公里）")
plt.grid(True)
plt.show()
best_pace = min(pace)
print(f"最快配速: {best_pace} 分钟/公里")

# 平均配速
avg_pace = sum(pace) / len(pace)
print(f"平均配速: {avg_pace:.2f} 分钟/公里")

# 进步了多少
improve = pace[0] - pace[-1]
print(f"从第一天到最后一天，配速快了 {improve:.2f} 分钟/公里")
```
